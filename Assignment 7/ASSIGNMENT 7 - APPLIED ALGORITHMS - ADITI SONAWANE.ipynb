```python
# PROBLEM 2
import heapq

class Node:
    def __init__(self, value, minimum, index):
        self.value = value
        self.minimum = minimum
        self.index = index
 
    def __lt__(self, other):
        return self.value < other.value

def minimum_range(lists): 
    maximum_element = 0
    range1 = [0, 10000]
 
    min_heap = []

    for i in range(len(lists)):
        if not lists[i]:    
            return -1, -1
        heapq.heappush(min_heap, Node(lists[i][0], i, 0))
        maximum_element = max(maximum_element, lists[i][0])
 
    while True:
 
        temp = heapq.heappop(min_heap)
 
        minimum_element = temp.value
        i = temp.minimum
        j = temp.index
 
        if maximum_element - minimum_element < range1[1] - range1[0]:
            range1 = (minimum_element, maximum_element)
 
        if j == len(lists[i]) - 1:
            return range1
 
        heapq.heappush(min_heap, Node(lists[i][j + 1], i, j + 1))
 
        maximum_element = max(maximum_element, lists[i][j + 1])

        
    return (minimum_element, maximum_element)

lists = [[ 3, 6, 8, 10, 15],[ 1, 5, 12 ],[ 4, 8, 15, 16 ],[ 2, 6 ]]
print("Output : The minimum range is",minimum_range(lists))
lists = [[ 2, 3, 4, 8, 10, 15 ],[ 1, 5, 12 ],[ 7, 8, 15, 16 ],[ 3, 6 ]]
print("Output : The minimum range is",minimum_range(lists))
```
