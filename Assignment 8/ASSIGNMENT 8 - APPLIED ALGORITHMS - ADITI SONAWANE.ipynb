# PROBLEM 1

```python
def mergesort(list):
    
    if len(list) <= 1:
        return list
    mid = len(list) // 2

    l = mergesort(list[:mid])
    r = mergesort(list[mid:])

    return merge(l, r)

def merge(l,r):
    result = []
    li = ri = 0

    for i in range(len(l) + len(r)):
        if li < len(l) and ri < len(r):
            if l[li] <= r[ri]:
                result.append(l[li])
                li += 1
            
            else:
                result.append(r[ri])
                ri += 1

        elif li == len(l):
            result.append(r[ri])
            ri += 1
    
        elif ri == len(r):
            result.append(l[li])
            li += 1
    return result

def maxEmployees(list1, list2):
    list1 = mergesort(list1)
    list2 = mergesort(list2)
    max = 0
    max_count = 0
    count = 0
    i = 0
    j = 0
    while i<len(list1) and j<len(list2):
        if list1[i]<=list2[j]:
            count+=1
            if count>max:
                max = count
                max_count = 1
            elif count == max:
                max_count = max_count + 1
            i+=1
        else:
            count-=1
            j+=1
    
    return (max, max_count)


 
#clock_in_times = [1,2]
#clock_out_times = [1,4]
#clock_in_times =[2,2,3,5,6,7,10]
#clock_out_times =[4,5,7,10,6,10,11] 
clock_in_times = [1, 2, 4, 7, 8, 12]
clock_out_times = [3, 7, 8, 12, 10, 15 ]
print("Output :",maxEmployees(clock_in_times,clock_out_times))
```

# PROBLEM 3

```python
def inplace(arr,i,j):
    
    pivot = arr[j]  
    mid = i - 1
    
    if i < j:  
        for k in range(i,j+1):
            if arr[k] <= pivot:
                
                mid += 1
                
                temp = arr[mid]
                arr[mid] = arr[k]
                arr[k] = temp
                
                 
        inplace(arr,i,mid-1)  
        inplace(arr,mid+1,j)  
    return arr

def quicksort(arr):
    
    return inplace(arr,0,len(arr)-1)



print(quicksort([4,3,1,2]))
print(quicksort([7,7,3,2,1]))
```

# PROBLEM 5

```python
# I am checking for the most overlapped string in the array using backtracking. 
# Assumption is that the result string length will always be greater than or equal to one of the two strings 
# that are being compared.This solution is optimal when most of the strings in array have very common prefix and suffixes.

def check_maximum_length(A,t,i):
    for j in range(len(A)):
        if i!=j:
            maximum_length=0
            for k in range(1,min(len(A[i]),len(A[j]))):
                if A[j][:k]==A[i][-k:]:
                    maximum_length=k
            t[(i,j)]=maximum_length
    return

def combine_string(array, temp_array, result):
    l = len(result)
    while array:
        
        temp_string,current_index,remaining_index=array.pop()
        #print("str",temp_string)
        #print("current",current_index)
        #print("other",remaining_index)
        if len(temp_string)<l:
            if not remaining_index:
                result=temp_string
                l=len(result)
            else:
                temp=[[temp_array[current_index,x],x] for x in remaining_index] 
                #print(temp)
                temp.sort()
                for i,j in temp:
                    final=temp_string+A[j][i:]
                    rindex=set(remaining_index)
                    rindex.remove(j)
                    array.append((final,j,rindex))
                    
    return result


def shortest_superstring(A):
    n = len(A)
    if n==1:
        return A[0]
    temporary_array={}
    i = 0
    while i<n:
        check_maximum_length(A,temporary_array,i)
        i=i+1
    if max(temporary_array.values())==0:
        return "".join(A)
    else:
        result="".join(A)
        l=len(result)
        stack=[]
        for i,j in enumerate(A):
            temp=set(range(n))
            temp.remove(i)
            stack.append((j,i,temp))
            result = combine_string(stack, temporary_array, result)
        
    
        return result
    
    
A = ['CATGC', 'CTAAGT', 'GCTA', 'TTCA', 'ATGCATC']
print("Output: The shortest superstring is =",shortest_superstring(A))
```
