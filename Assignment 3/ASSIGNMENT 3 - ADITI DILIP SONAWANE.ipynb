```python
# PROBLEM 1

countMV = 0
countCP = 0
countLS = 0

def count(i):
    global countMV
    global countCP
    global countLS
    if i == 'ls':
        countLS = countLS + 1
    elif i == 'cp':
        countCP = countCP + 1
    elif i == 'mv':
        countMV = countMV + 1

def commands_count (commands) :
    
    for i in range(0,len(commands)):
        if '!' in commands[i]:
            temp = commands[i].replace('!'," ")
            temp = int(temp)
            commands[i] = commands[temp-1]
            #print(commands[i])
            count(commands[i])
        else:
            count(commands[i])   
    
    result = [countCP,countLS,countMV]
    return result
        
commands = [ "ls", "cp", "mv", "mv", "mv", "!1", "!3", "!6"]
print(commands_count(commands))
```

```python
# PROBLEM 3
import math

import random



class Node:
    def __init__(self, level, word):
        self.word = word
        self.next = [None]*(level+1)
        
        
        
        
class LookUpSkipList:
    def __init__(self,words,probability):
        self.checkHeight = int(math.log(len(words),2))
        self.head = self.new_node(self.checkHeight,-1)
        self.level = 0
        self.probability = probability
        
    def new_node(self, level, word):
        newnode = Node(level, word)
        return newnode
    
    
    
    
    def Level(self):
        level = 0
        while random.random()<self.probability and level<self.checkHeight:level +=1
        return level
    
    
    
    
    def Search_List(self,word):
        temp = self.head
        for i in range(self.level,-1,-1):
            while temp.next[i] and temp.next[i].word < word:
                temp = temp.next[i]
        temp = temp.next[0]
        if temp and temp.word == word:
            print("True")
            return True
        else:
            print("Word Not found, inserted in the list")
            self.insert(word)
            
        
        
        
        
            
            
            
    def insert(self,word):
        update = [None]*(self.checkHeight+1)
        temp = self.head
        
        
        for i in range(self.level,-1,-1):
            
            while temp.next[i] and temp.next[i].word < word:
                temp = temp.next[i]
                
            update[i] = temp
            
            
            
        temp = temp.next[0]
        if temp == None or temp.word != word:
            x = self.Level()
            
            
            if x > self.level:
                for i in range(self.level+1,x+1):
                    update[i] = self.head
                self.level = x
            new_node = self.new_node(x,word)
            for i in range(x+1):
                new_node.next[i] = update[i].next[i]
                update[i].next[i] = new_node
    
    
    
    
    
    def print_list(self):
        head = self.head
        
        for i in range(0,self.level+1):
            print('Level',i,'-')
            temp = head.next[i]
            while (temp != None):
                print(temp.word)
                temp = temp.next[i]
                
                
                
                
                

words = ['iub','usa','there','sort','god']
skiplist = LookUpSkipList(words,0.6)
for i in words:
    skiplist.insert(i)
    
    
print("After inserting initial list :")
skiplist.print_list()

skiplist.insert('word')


print("After inserting new word - 'word' :")


skiplist.print_list()

print("Search result for 'iub' :")

skiplist.Search_List('iub')
print("Search result for 'present' :")
skiplist.Search_List('present')
print("After inserting word 'present' :")
skiplist.print_list()
```

```python
# PROBLEM 4

def predict_winner (s) :
    countA = 0
    countO = 0
    for i in s:
        if i=='A':
            if countA>=0:
                countO = countO - 1
            countA = countA + 1
        if i=='O':
            if countO>=0:
                countA = countA - 1
            countO = countO + 1
    if countA == countO:
        if s[0]=="A":
            return "apple"
        else:
            return "orange"
    if countA>countO:
        return "apple"
    return "orange"
            
            
print("Output -",predict_winner("AOOAOAOAAAAOOAOOAAOA"))
print("Output -",predict_winner("AAOOO"))
print("Output -",predict_winner("AOO"))
print("Output -",predict_winner("AAOOOA"))
print("Output -",predict_winner("AO"))
```
